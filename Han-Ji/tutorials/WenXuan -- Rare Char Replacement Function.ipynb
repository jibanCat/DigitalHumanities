# WenXuan -- Rare Char Replacement Function

```python
from WenXuan import WenXuan

wenshuan = WenXuan("2018-07-10", "MF")
wenshuan.load_htmls()
```

## Replacement

```python
# if the wenshuan_rare_char.json does not exist
import os
if not os.path.exists("{}_rare_char.json".format(wenshuan.bookname)):
    driver_path = '/Users/cicero/Documents/Physics/PYTHON/Scraping/chromedriver'
    wenshuan.extract_rare_chars(driver_path) # this line would take a very long time, be careful before you execute it
    wenshuan.write_rare_chars()
```

What `extract_rare_chars` do is writing each html into a temp file `temp.html` with sinca [normalization javascript API](http://char.iis.sinica.edu.tw/API/ics.js). And then, open the html with `selenium` and scrape the rare chars identified by the API. Finally, using the sinica normalization search to identify the unicode of the rare chars.

Because sinica normalization API would block out the massive amount of search, sometime the `extract_rare_chars` function would appear some errors.

```python
# if the wenshuan_rare_char.json exist
wenshuan.update_rare_chars()
```

Yeah, üëÜthis line update the html source page to recover the UNICODE of the rare characters. Therefore, rare char components in the `self.flat_bodies` would have been converted to UNICODEs.

## Save to CSV

```python
import pandas as pd
df = pd.DataFrame({key : values
    for rare_chars in wenshuan.flat_rare_chars
    for key,values in rare_chars.items()
}).T
df.columns = ['UNICODE', 'UTF-8']
df.to_csv("WenXuan_rare_char.csv")
```
